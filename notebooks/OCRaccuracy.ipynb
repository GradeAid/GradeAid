```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
```

```python
x=Image("/content/drive/MyDrive/dataset/Cat/Cat2.jpg")
```

```python
x
```

```python
rcParams["figure.figsize"] = 10, 10
reader = easyocr.Reader(["en"])
output = reader.readtext(cv2.imread("/content/drive/MyDrive/dataset/Cat/Cat2.jpg"))
text = ""
for i in output:
    text += i[1] + " "
print(text)
```

```python
text=text.lower()
text=text.replace("_"," ")
text=text.split(" ")

```

```python
file1 = open('/content/drive/MyDrive/dataset/Cat/Cat.txt', "r")
FileContent = file1.read()
```

```python
original=FileContent.lower()
original=original.split(" ")
```

```python
intersect = [value for value in text if value in original]
```

```python
print("Accuracy: "+str(len(intersect)/len(original)*100)+"%")
```
